```python
print("ğŸ¯ HALAL CHECKER - INTERFACE SEDERHANA")
print("=" * 60)

# Import dari file deployment yang sudah ada
import sys
sys.path.append('/content/drive/MyDrive/halal_classifier_deployment')

from halal_classifier_deployment import HalalClassifier

# Load classifier
classifier = HalalClassifier(
    model_path='halal_classification_linear_svm.pkl',
    vectorizer_path='tfidf_vectorizer.pkl',
    metadata_path='model_metadata.pkl'
)

print("\nğŸ” MASUKKAN BAHAN PRODUK UNTUK DICEK")
print("Contoh: wheat flour water salt yeast sugar")
print("Ketik 'selesai' untuk berhenti")
print("-" * 50)

while True:
    # Input dari user
    bahan = input("\nğŸ§ª Masukkan daftar bahan: ")

    if bahan.lower() in ['selesai', 'stop', 'exit']:
        print("Terima kasih telah menggunakan Halal Checker! ğŸ‘‹")
        break

    if not bahan.strip():
        print("âŒ Mohon masukkan bahan yang valid")
        continue

    # Predict
    result = classifier.predict(bahan)

    # Tampilkan hasil
    print("\n" + "=" * 50)
    print("ğŸ“Š HASIL ANALISIS HALAL/HARAM")
    print("=" * 50)

    if result['status'] == 'HALAL':
        print("âœ… STATUS: HALAL ğŸŸ¢")
        print("   Produk ini aman dikonsumsi")
    elif result['status'] == 'HARAM':
        print("âŒ STATUS: HARAM ğŸ”´")
        print("   Produk ini TIDAK halal")
    else:
        print("âš ï¸ STATUS: TIDAK PASTI ğŸŸ¡")
        print("   Perlu pengecekan lebih lanjut")

    print(f"ğŸ¯ Tingkat Keyakinan: {result['confidence']:.2%}")
    print(f"ğŸ“ˆ Detail Probabilitas:")
    print(f"   â€¢ HALAL: {result['probabilities']['halal']:.2%}")
    print(f"   â€¢ HARAM: {result['probabilities']['haram']:.2%}")
    print(f"ğŸ”§ Jumlah bahan terdeteksi: {result['ingredients_count']}")
    print("=" * 50)
```
